```
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as solveur
```

```
Ca0 = 1.5 #M
Cb0 = 2 #M
Vdot = 10 #l/min

Fa0 = Ca0 * Vdot
Fb0 = Cb0 * Vdot
Fc0 = 0
Fd0 = 0
Fe0 = 0
Ff0 = 0 
```

```
def equations(variables, t, ):
  
  Fa, Fb, Fc, Fd, Fe, Ff = variables
  
  r1 = 0.25*(Fa/Vdot)*(Fb/Vdot)
  r2 = 0.1*(Fa/Vdot)*(Fd/Vdot)
  r3 = 5*(Fb/Vdot)**2*(Fc/Vdot)
  
  dFadt = -r1-3*r2
  dFbdt = -2*r1-r3
  dFcdt = r1 + r2-2*r3
  dFddt = r1-2*r2+r3
  dFedt = r2
  dFfdt = r3
 
  return dFadt, dFbdt, dFcdt, dFddt, dFedt, dFfdt
  
```

```
conditions_initiales = [Fa0,Fb0,Fc0,Fd0,Fe0,Ff0 ]
```

```
t = np.linspace(0,50,51) #bornes d'int√©gration
resultats =solveur.odeint(equations,conditions_initiales, t)
```

```
plt.plot(t,resultats)
plt.xlabel("V [l]")
plt.ylabel("Fi [mol/min]")
plt.legend(["Fa","Fb","Fc","Fd","Fe","Ff"])
plt.show
```
